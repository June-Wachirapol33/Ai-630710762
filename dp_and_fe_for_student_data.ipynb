**3.2 การทำให้ข้อมูลเป็นมาตรฐาน**

```python
def standardize_faculty(name):
        if pd.isna(name):
            return 'ไม่ระบุ'
        name_lower = str(name).lower().strip()
        if 'วิทยาศาสตร์' in name_lower or 'วิทย์' in name_lower or 'science' in name_lower or 'sci' in name_lower:
            return 'วิทยาศาสตร์'
        elif 'ศิลปศาสตร์' in name_lower or 'ศิลป์' in name_lower or 'arts' in name_lower or 'liberal arts' in name_lower:
            return 'ศิลปศาสตร์'
        elif 'วิศวะ' in name_lower or 'วิศวกรรม' in name_lower or 'engineering' in name_lower:
            return 'วิศวกรรมศาสตร์'
        else:
            return 'อื่นๆ'
df['คณะ_มาตรฐาน'] = df['คณะ'].apply(standardize_faculty)
df[['ชื่อ', 'คณะ', 'คณะ_มาตรฐาน']].head()
```

คำอธิบาย Code

* ทำให้ข้อมูลในคอลัมน์ 'คณะ' เป็นมาตรฐาน (Standardization)


**3.3 การจัดการค่าผิดปกติ**

```python
df['อายุ'] = df['อายุ'].fillna(df['อายุ'].mean())
Q1 = df['อายุ'].quantile(0.25)
Q3 = df['อายุ'].quantile(0.75)
IQR = Q3 - Q1
```

คำอธิบาย Code
* จัดการค่าผิดปกติ (Outliers) โดยใช้วิธี IQR
* ค่า Q1, Q3, และ IQR ที่คำนวณได้ในเซลล์นี้จะถูกนำไปใช้ในเซลล์ถัดไปเพื่อกำหนดขอบเขตในการตรวจหาค่าผิดปกติ

```python
# กำหนดขอบเขตบนและล่าง
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

คำอธิบาย Code
* คำนวณขอบเขตบนและล่างที่ใช้สำหรับตรวจหาค่าผิดปกติ (Outliers) ในคอลัมน์ 'อายุ

```python
# กรองค่าผิดปกติ
df_filtered = df[(df['อายุ'] >= lower_bound) & (df['อายุ'] <= upper_bound)]
```

คำอธิบาย Code
*  กรองค่าผิดปกติ (Outliers) ในคอลัมน์ 'อายุ'

```python
print("\n--- ข้อมูลหลังกรองค่าผิดปกติ ---")
print(df_filtered[['ชื่อ', 'อายุ']])
print(f"\nขอบเขตล่าง: {lower_bound:.2f}")
print(f"ขอบเขตบน: {upper_bound:.2f}")
print(f"\nค่าเฉลี่ยอายุหลังกรองค่าผิดปกติ: {df_filtered['อายุ'].mean():.2f}")
```

คำอธิบาย Code
* แสดงผลลัพธ์หลังจากที่เราได้กรองค่าผิดปกติ (Outliers) ในคอลัมน์ 'อายุ'

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'])
```

* แปลงคอลัมน์ 'วันที่ลงทะเบียน' ให้เป็นชนิดข้อมูลแบบวันเวลา (Datetime)

```python
df['ปีที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.year
df['เดือนที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.month
df['วันในสัปดาห์ที่ลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.day_name()
def get_season(month):
        if month in [3, 4, 5]:
            return 'ร้อน'
        elif month in [6, 7, 8, 9, 10]:
            return 'ฝน'
        else:
            return 'หนาว'
df['ฤดูที่ลงทะเบียน'] = df['เดือนที่ลงทะเบียน'].apply(get_season)
print(df[['ชื่อ', 'วันที่ลงทะเบียน', 'ปีที่ลงทะเบียน', 'เดือนที่ลงทะเบียน', 'วันในสัปดาห์ที่ลงทะเบียน', 'ฤดูที่ลงทะเบียน']].head())
df[['ชื่อ', 'วันที่ลงทะเบียน', 'ปีที่ลงทะเบียน', 'เดือนที่ลงทะเบียน', 'วันในสัปดาห์ที่ลงทะเบียน', 'ฤดูที่ลงทะเบียน']].head()
```

**4.3 การรวม Features**

```python
df['ส่วนสูง_cm'] = df['ส่วนสูง_cm'].fillna(df['ส่วนสูง_cm'].mean())
df['น้ำหนัก_kg'] = df['น้ำหนัก_kg'].fillna(df['น้ำหนัก_kg'].mean())
```

คำอธิบาย Code
*  เติมค่าที่หายไป (missing values) ในคอลัมน์ 'ส่วนสูง_cm' และ 'น้ำหนัก_kg'

```python
# สูตร: BMI = น้ำหนัก (kg) / [ส่วนสูง (m)]^2
df['ส่วนสูง_m'] = df['ส่วนสูง_cm'] / 100
df['BMI'] = df['น้ำหนัก_kg'] / (df['ส่วนสูง_m'] ** 2)
```

คำอธิบาย Code
* คำนวณค่าดัชนีมวลกาย (BMI) และสร้างเป็น Feature ใหม่

```python
df['สุขภาพ'] = pd.cut(df['BMI'],
                           bins=[0, 18.5, 25, 30, 100],
                           labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'],
                           right=False) # ใช้ right=False เพื่อรวมค่าต่ำสุดในแต่ละ bin
```

คำอธิบาย Code
* สร้าง Feature ประเภท Categorical ใหม่ 'สุขภาพ' โดยอ้างอิงจากค่า BMI ที่คำนวณได้

```python
print("\n--- ผลลัพธ์หลังการสร้าง Feature BMI และสุขภาพ ---")
print(df[['ชื่อ', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'BMI', 'สุขภาพ']].head(12))
df[['ชื่อ', 'ส่วนสูง_cm', 'น้ำหนัก_kg', 'BMI', 'สุขภาพ']].head()
```

คำอธิบาย Code
* แสดงผลลัพธ์หลังจากที่เราได้สร้าง Feature ใหม่คือ BMI และสุขภาพ
* ช่วยให้เราตรวจสอบได้อย่างรวดเร็วว่าการคำนวณ BMI และการจัดกลุ่มสถานะสุขภาพถูกต้อง
